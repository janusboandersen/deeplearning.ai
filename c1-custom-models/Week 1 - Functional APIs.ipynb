# Functional API in Tensorflow/Keras

Allows more control of the structure of the networks, e.g.
- Non-sequential structure
- Repeated / re-used modules
- Separate channels with subsequent concatenation


```python
from tensorflow.keras.datasets import mnist
```

```python
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.models import Model, Sequential
from tensorflow.python.keras.utils.vis_utils import plot_model
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
N_train = x_train.shape[0]
N_test = x_test.shape[0]
```

```python
X = x_train.reshape((N_train, 28, 28))/255.0
y = y_train
```

```python
X.shape
```

```python
def build_func_model():
    input = Input(shape=(28,28))
    x = Flatten()(input)
    x = Dense(128, activation='relu')(x)
    predictions = Dense(10, activation='softmax')(x)
    func_model = Model(inputs=input, outputs=predictions)
    return func_model
```

```python
model = build_func_model()
model.summary()
```

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='sgd',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X,
    y,
    epochs=10
)
```

### Regression example

Using the Sequential API

```python
xs = np.array([-1, 0, 1, 2, 3, 4, 5, 6, 7], dtype=float)
ys = 2*xs - 1
```

```python
regression_layer = Dense(units=1, input_shape=[1])
```

```python
reg_model = Sequential([regression_layer])
```

```python
reg_model.compile(
    optimizer='sgd',
    loss='mean_squared_error'
)
```

```python
reg_model.summary()
```

```python
reg_model.fit(xs, ys, epochs=500)
```

```python
reg_model.predict([10])
```
